```python
transliterated_sentences = []
bulgarian_sentences = []
with open('../drive/MyDrive/bg_sentences.txt', 'r') as file:
  for line in file:
    if not line.startswith("\n"):
      temp = cyrtranslit.to_latin(line.strip(), "bg") + "\n"
      bulgarian_sentences.append(line.strip())
      transliterated_sentences.append(temp)
```

```python
tag_bg = []
tag_tl = []
pos_tagger = POSTagger()
for i in range(len(bulgarian_sentences)):
  temp_tag_bg = pos_tagger.generate_tags(bulgarian_sentences[i])
  tag_bg.append(temp_tag_bg)

  temp_tag_tl = pos_tagger.generate_tags(transliterated_sentences[i])
  tag_tl.append(temp_tag_tl)
```

```python
filtered_tag_bg = []
filtered_tag_tl = []
wrong_sentences = []
for i in range(len(tag_bg)):
  if len(tag_bg[i]) == len(tag_tl[i]):
    filtered_tag_bg.append(tag_bg[i])
    filtered_tag_tl.append(tag_tl[i])
  else:
    wrong_sentences.append(i)
```

```python
flat_tag_bg = []
flat_tag_tl = []
for i in range(len(filtered_tag_bg)):
    flat_tag_bg.extend(filtered_tag_bg[i])
    flat_tag_tl.extend(filtered_tag_tl[i])
```

```python
from sklearn.metrics import confusion_matrix
labels = list(set(flat_tag_bg))
cm = confusion_matrix(flat_tag_bg, flat_tag_tl, labels = labels, normalize = 'true')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 6))
sns.heatmap(cm,cmap='YlGnBu', xticklabels=labels, yticklabels=labels)
plt.ylabel('Prediction',fontsize=12)
plt.xlabel('Actual',fontsize=12)
plt.title('Confusion Matrix',fontsize=16)
plt.show()
```
